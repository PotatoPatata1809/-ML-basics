```python
import pandas as pd

df = pd.read_csv("/kaggle/input/student-performance-dataset/students performance.csv")
df.head()
```

```python

df.info()
```

```python
num_cols = ['math score', 'reading score', 'writing score']
df[num_cols].describe()
```

```python
print("Mean:\n", df[num_cols].mean())
print("\nMedian:\n", df[num_cols].median())
print("\nMode:\n", df[num_cols].mode())
```

```python
print("Minimum:\n", df[num_cols].min())
print("\nMaximum:\n", df[num_cols].max())
print("\nSum:\n", df[num_cols].sum())
print("\nVariance:\n", df[num_cols].var())
print("\nStandard Deviation:\n", df[num_cols].std())
```

```python
Q1 = df[num_cols].quantile(0.25)
Q2 = df[num_cols].quantile(0.50)
Q3 = df[num_cols].quantile(0.75)

print("Q1 (25%):\n", Q1)
print("\nQ2 (Median):\n", Q2)
print("\nQ3 (75%):\n", Q3)
```

```python
# Correlation
correlation = df[num_cols].corr()
print("Correlation Matrix:\n", correlation)
```

```python
# Covariance
covariance = df[num_cols].cov()
print("Covariance Matrix:\n", covariance)
```

```python
plt.hist(df['math score'], bins=10)
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.title("Histogram of Math Scores")
plt.show()
```

```python
plt.boxplot(df['math score'])
plt.title("Boxplot of Math Scores")
plt.ylabel("Score")
plt.show()
```
